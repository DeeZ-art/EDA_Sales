```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import glob
import os
all_files = glob.glob(os.path.join(r"C:\Users\Deepak Chaudhary\Downloads\Sales_Dataset", "*.csv"))
Data = pd.concat((pd.read_csv(f) for f in all_files), ignore_index=True)
```

## Removing the duplicate rows because of the low variance

```python
Data.drop_duplicates()
```

## Removing all the null values as well

```python
Data = Data.dropna()
Data.isnull().sum()
```

## Removing all the empty rows

```python
Data = Data.replace('', np.nan).dropna()
Data
```

```python
prod_uni = Data.Product.unique().tolist()
length = [i for i in range(len(prod_uni))]
Data = Data.reset_index()
```

## Converting the CategoricalCol(Product) to NumericCol(CatProd)

```python
res = {prod_uni[i]: length[i] for i in range(len(prod_uni))}
lis = [res[i] for i in Data.Product]
lis=pd.DataFrame(lis, columns=['CatProd'])
Data.insert(2, "CatProd", lis["CatProd"])
Data
```

## Converting the Order Date column into Day, Month, Year and Time columns separately

```python
Data[['Date','Time']]=Data['Order Date'].str.split(' ',expand=True)
Data
```

```python
Data.drop(['index'], axis=1,inplace = True)
```


## Creating a new column and storing the Price data as rounded off in integer format

```python
Data.drop(Data[Data['Price Each'] == 'Price Each'].index,inplace = True)
```

```python
Data = Data.reset_index()
Data.drop(['index'], axis=1,inplace = True)
Data
```

```python
Data['Price Each']=Data['Price Each'].astype('float64')
```

```python
y = list(map(round,Data['Price Each'].tolist()))
y = pd.DataFrame(y,columns = ['roundPrice'])
Data.insert(5, "RoundPrice", y["roundPrice"])
```

## Creating a new column and storing the PIN Codes from the Address column

```python
Pin_Code= [data[-5:] for data in Data['Purchase Address'].tolist()]
Pin_Code = pd.DataFrame(Pin_Code,columns = ['Pin_Code'])
Data.insert(8,'Pin_Code',Pin_Code["Pin_Code"])
```

```python
Data
```

## Creating a new Column named City from Purchase Address

```python
x = Data['Purchase Address'].tolist()
y = [data[:-10].split(", ") for data in x]
res = [data[1] for data in y]
res = pd.DataFrame(res,columns = ['City'])
Data.insert(9,'City',res['City'])
```

```python
Data
```

## Grouping Product column into different categories such as Phones, Accessories etc.

```python
Data.Product.unique().tolist()
Prod = Data.Product.tolist()
Cat = []
for data in Prod:
    if data in ['Wired Headphones']: Cat.append('Wired_Headpones')
    elif data in ['Bose SoundSport Headphones','Apple Airpods Headphones']: Cat.append('Wireless_Headphones')
    elif data in ['USB-C Charging Cable','Lightning Charging Cable']: Cat.append('Charging_Cables')
    elif data in ['Google Phone','iPhone','Vareebadd Phone']: Cat.append('Phones')
    elif data in ['Macbook Pro Laptop','ThinkPad Laptop']: Cat.append('Laptops')
    elif data in ['27in 4K Gaming Monitor','27in FHD Monitor','34in Ultrawide Monitor','20in Monitor']: Cat.append('Monitor')
    elif data in ['Flatscreen TV']: Cat.append('TV')
    elif data in ['AA Batteries (4-pack)','AAA Batteries (4-pack)']: Cat.append('Batteries')
    elif data in ['LG Dryer']: Cat.append('Dryer')
    elif data in ['LG Washing Machine']: Cat.append('Washing_Machine')
    else: Cat.append('Others')
Cat = pd.DataFrame(Cat,columns = ['Category'])
Data.insert(3,'Category',Cat['Category'])
```

## What was the best month for sales? How much was earned that month?

```python
Data[['Month','Day','Year']]=Data['Date'].str.split('/',expand=True)
Data
```

```python
plt.rcParams["axes.labelsize"] = 15
plt.rcParams['figure.figsize']=[20,6]
Month = Data.Month.unique().tolist()

price = [Data.loc[Data[Data.Month == data].index.values,'Price Each'].sum() for data in Month]
Month,price = zip(*sorted(zip(Month,price)))
sns.barplot(x = ['Jan','Feb','March','April','May','June','July','Aug','Sept','Oct','Nov','Dec'],y = list(price),palette = 'magma_r')
plt.grid()
plt.show()
```

## The best month for sale was December and earnings were around 4600000$
